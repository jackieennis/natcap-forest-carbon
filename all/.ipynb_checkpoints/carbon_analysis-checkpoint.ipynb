## Relationship analysis - correlations, plots, breakdowns
### Dataframe cleaning & preprocessing

```python
# Load in previously saved dataframes
carbon_5m_merged = pd.read_csv(os.path.join(OUTPUT_DIR, "carbon_5m_merged.csv"))
```

```python
carbon_5m_merged = carbon_5m_merged[['lat_carbon', 'long_carbon', 'carbon',
       'pixel_id', 'mins_to_market', 'market_influence', 'population_density',
        'infant_mortality_rate', 'nighttime_lights', 'pasture_land', 'soil_water_capacity']]
    
print(carbon_5m_merged.columns)
print(len(carbon_5m_merged))
carbon_5m_merged.head()
```

```python
# Keep track of all metrics we're analyzing
signals = ['mins_to_market', 'market_influence', 'population_density',
        'infant_mortality_rate', 'nighttime_lights', 'pasture_land', 'soil_water_capacity']
```

Clean up null or incoherent values.

```python
print("Number of records before cleaning:", len(carbon_5m_merged))

# We're only interested in high-carbon areas so remove low-carbon rows
carbon_5m_merged = carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['carbon'] < 100].index)

# For most other social signals, ignore null and negative values 
carbon_5m_merged = carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['mins_to_market'] < 0].index)

carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['market_influence'] < 0].index, 
    inplace=True)

carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['population_density'] < 0].index, 
    inplace=True)


carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['infant_mortality_rate'] < 0].index, 
    inplace=True)

carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['nighttime_lights'] < 0].index, 
    inplace=True)

carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['pasture_land'] < 0].index, 
    inplace=True)

carbon_5m_merged.drop(
    carbon_5m_merged[carbon_5m_merged['soil_water_capacity'] < 0].index, 
    inplace=True)

carbon_5m_merged.head()

print("Number of records after cleaning:", len(carbon_5m_merged))
```

### Looking for correlations

Is the correlation between covariates better than random?

```python
from scipy.stats import pearsonr

# Is the correlation better than random?
rand_corr = 0

for i in range(0, 100):
    carbon_5m_merged['rand'] = np.random.random(carbon_5m_merged.shape[0])
    new_r = pearsonr(carbon_5m_merged['carbon'], carbon_5m_merged['rand'])
    rand_corr += new_r[0]

rand_corr = rand_corr / 100

print("Correlation between carbon and random:", rand_corr)

for signal in signals:
    corr = pearsonr(carbon_5m_merged['carbon'], carbon_5m_merged[signal])
    print("Stats for", signal, ":", corr)
```

```python
# Print correlation matrix
import seaborn as sns
corr = carbon_5m_merged.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

## Plotting covariate relationships

```python
# Add log scale for carbon values
carbon_5m_merged["log_carbon"] = np.log(carbon_5m_merged['carbon'])
carbon_5m_merged.head()
```

```python
for signal in signals: 
    sample = carbon_5m_merged.sample(frac=0.1, replace=False, random_state=1)
    
    ylims = (0, 9)
    
    if signal=="mins_to_market":
        xlims = (0,2500)
    elif signal=="market_influence":
        xlims = (0,5000)
        ylims = (3, 8)
    elif signal=="population_density":
        xlims = (0,450)
        ylims = (4, 8)
    elif signal=="infant_mortality_rate":
        xlims = (0,80)
        ylims = (4, 8)
    elif signal=="nighttime_lights":
        xlims = (0,15)
        ylims = (4, 8)
    elif signal=="soil_water_capacity":
        xlims = (0,60)
        ylims = (4, 7.5)
    elif signal=="pasture_land":
        xlims = (0,100)
        ylims = (4, 7.5)

    print("Plotting KDE...")
    print(datetime.datetime.now())
    g = sns.jointplot(signal, "log_carbon", data=sample, 
                      kind="kde", space=0, color="g", 
                      xlim=xlims, ylim=ylims)
```

```python
sample = carbon_5m_merged.sample(frac=0.1, replace=False, random_state=1)

xlims = (0,0.8)
ylims = (4, 7.5)

print("Plotting KDE...")
print(datetime.datetime.now())
g = sns.jointplot("pasture_land", "log_carbon", data=sample, 
    kind="kde", space=0, color="g", 
    xlim=xlims, ylim=ylims)
```
